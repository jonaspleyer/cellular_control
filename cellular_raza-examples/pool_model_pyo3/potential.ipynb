```python
import numpy as np
import matplotlib.pyplot as plt

def potential(sigma):
    q = 0.2
    modifier = (1.0 + q) / (q + sigma)
    return modifier * (sigma < 1.0)

fig, ax = plt.subplots()

x = np.linspace(0, 1.5, 200)
y = potential(x)

ax.set_title("Relative Interaction Strength")
ax.set_xlabel('Radial Coordinate $\\sigma$')
ax.set_ylabel('Strength $\\epsilon^{-1}||\\vec{F}||$')
ax.plot(x, y, color="black")
fig.savefig("interaciton_potential.png")
```
