```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def force(x, p):
    radius_ext, radius_int, interaction_range, potential_strength = p
    sigma = x / (radius_ext + radius_int)
    alpha = 3/2*(interaction_range/(radius_ext + radius_int))
    form = (3*(sigma-1)**2 - 2*alpha*(sigma-1))*3/alpha**2
    return potential_strength * form
```

```python
radius_int = 1.0
radius_ext = 1.2
interaction_range = 2.0
potential_strength = 1.0

bound = 1.0

p = (radius_ext, radius_int, interaction_range, potential_strength)

x = np.linspace(0.01, radius_int + radius_ext + interaction_range)
y = np.clip(force(x, p), -bound, bound)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, 0*x, color="k", alpha=0.3)
ax.axvline(radius_int, color="k")
ax.axvline(radius_int+radius_ext, color="r")
plt.show()
```
